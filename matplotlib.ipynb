### Data visualization practice

```python
from matplotlib.backends.backend_agg import FigureCanvasAgg as FigureCanvas  #import FigureCanvas
```

```python
from matplotlib.figure import Figure #import Figure Artist
```

```python
fig = Figure()
```

```python
canavas = FigureCanvas(fig)
```

```python
#create 1000 random numbers using numpy
import numpy as np
x = np.random.randn(1000)

ax = fig.add_subplot(111) #create an axes artist
ax.hist(x, 100)  #generate a histogram of 1000 numbers

#add a title to the fig and save it
ax.set_title("Normal distribution with $\mu=0, \sigma=1$")
fig.savefig('matplotlib_histrogram.png')
```

### Scripting method for pyplot

All the axes, fig methods are automatically imported by pyplot in script layer

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline

x = np.random.randn(1000)
plt.hist(x, 100)
plt.title("Normal distribution with $\mu=0, \sigma=1$")
plt.savefig("matplotlib_histogram.png")
plt.show()
```
